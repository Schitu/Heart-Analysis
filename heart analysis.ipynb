#  Plot Continuous Variables

```python
data.columns
```

```python
categorical_values=[]
continuos_values=[]
for coloumn in data.columns:
    if data[coloumn].nunique()<10:
        categorical_values.append(coloumn)
    else:
        continuos_values.append(coloumn)
```

```python
categorical_values
```

```python
continuos_values
```

```python
data.hist(categorical_values, figsize=(15,8))
plt.tight_layout()
plt.show()
```

```python
data.hist(continuos_values, figsize=(15,8))
plt.tight_layout()
plt.show()
```
